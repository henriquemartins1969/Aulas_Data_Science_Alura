Para isso, podemos executar um teste que irá, de fato, testar se essa hipótese é válida ou não. Nesse caso, vamos executar outro tipo de teste. Poderíamos usar o ztest(), assim como fizemos nos passos anteriores, mas esse caso apresenta dois pontos diferentes:

Não temos informações sobre a população;
E temos uma amostra pequena, com menos de 30 observações.
Sendo assim, vamos usar o teste T. Esse teste faz basicamente a mesma coisa que o ztest(), porém, é mais apropriado para essas situações específicas.

Ao examinar o gráfico, surge a questão: há uma diferença substancial nesse comportamento? É realmente o caso de que quem foi exposto aos anúncios no e-commerce gastou mais? Para determinar se há uma diferença significativa entre esses dois grupos, é necessário conduzir um teste de independência.

####Teste de independência
As pessoas que compraram com anúncio são distintas e independentes das que compraram sem anúncio. Na célula seguinte, importamos o teste de independência do ***scipy.stats*** com ***from scipy.stats import ttest_ind***.

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind (vendas_com_propaganda, vendas_sem_propaganda)
```

```python
#Hipóteses desse teste de amostras independentes:

#H0: média sem = média com

#H1: média sem < média com
```

```python
ttest_ind (vendas_com_propaganda, vendas_sem_propaganda, alternative = "greater")
```

```python
stats, p_valor = ttest_ind (vendas_com_propaganda, vendas_sem_propaganda, alternative = "greater")
```

```python
print(p_valor)
```

####Quando observamos o p_valor, visualizamos que ele é um número muito alto, 48%. A probabilidade é muito alta no cenário de H₀, ou seja, não temos evidências de rejeitar o que está escrito em H₀, que, na verdade, não tem efeito nenhum. A média de quem compra com propaganda não tem muita diferença de quem compra sem propaganda.

Podemos usar o recurso de estabelecer o alpha=0.05, o nível de confiança de 95%, e esse p_valor.

```python
alpha = 0.05
if p_valor < alpha:
    print("Rejeitar a hipótese nula: Há evidências de que a propaganda tem um efeito positivo nas vendas.")
else:
    print("Não rejeita hipótese nula: Não há evidências suficientes para afirmar um efeito positivo nas vendas devido à propaganda.")
```
