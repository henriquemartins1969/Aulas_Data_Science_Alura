#Estabelecendo um intervelo de confiança

```python
# importando o stats
from scipy import stats
```

```python
# levar em consideração essa amostra
dados_vida_lampada

# nível de confiança definido
confianca = 0.95
```

```python
#informações para o IC

media = dados_vida_lampada['duracao'].mean()
desvio_padrao_amostral = dados_vida_lampada['duracao'].std()
tamanho_amostra = len(dados_vida_lampada)
```

```python
intervalo_confianca = stats.norm.interval(confianca)
```

```python
# Calculando intervalo de confiança

intervalo_confianca = stats.norm.interval(confianca,
                                       loc=media ,
                                       scale=desvio_padrao_amostral/np.sqrt(tamanho_amostra))
```

```python
# Visualizando os resultados

print("IC (95%):", intervalo_confianca)
```

```python
# Estabelecidos o nível de confiança em 95% e o desvio padrão populacional em 2.65,
# verifique se a média da amostra está definida dentro do intervalo de confiança da hipótese nula.
```

```python
import numpy as np

# Valores previamente especificados
confianca = 0.95
desvio_padrao_populacional = 2.65

# Outros valores para calcular o intervalo
tamanho_amostra = len(df_techtaste)
media_nula = 30

from scipy import stats
intervalo_confianca = stats.norm.interval(confianca,
                                       loc= media_nula,
                                       scale=(desvio_padrao_populacional)/np.sqrt(tamanho_amostra))

print(f'Intervalo de Confiança ({confianca*100}%): {intervalo_confianca}')
```

###Aplicando o teste T
No nosso notebook, há uma lista que armazena o tempo de duração para cada suporte. O primeiro suporte, por exemplo, durou 28 minutos; o segundo, 32 minutos; e assim sucessivamente. Tudo está armazenado na variável tempo_resposta.

```python
# Tempo de resposta em minutos para 25 solicitações de suporte
tempo_resposta = [28, 32, 29, 31, 30, 33, 28, 30, 31, 29,
                  30, 32, 29, 31, 30, 33, 28, 30, 31, 29,
                  30, 32, 24, 29, 30]
```

```python
#H₀: μ = 30
#H₁: μ < 30
```

###Analisando a média de tempo coletada
Como nos referimos a uma lista, vamos utilizar o método mean() do NumPy, que apelidamos como np. Portanto, vamos chamar np.mean() e passar tempo_resposta entre parênteses.

```python
# Média coletada do tempo
np.mean(tempo_resposta)
```

###Validando a hipótese
Para isso, podemos executar um teste que irá, de fato, testar se essa hipótese é válida ou não. Nesse caso, vamos executar outro tipo de teste. Poderíamos usar o ztest(), assim como fizemos nos passos anteriores, mas esse caso apresenta dois pontos diferentes:

Não temos informações sobre a população;
E temos uma amostra pequena, com menos de 30 observações.
Sendo assim, vamos usar o teste T. Esse teste faz basicamente a mesma coisa que o ztest(), porém, é mais apropriado para essas situações específicas.

Começaremos importando o método ttest_1samp do scipy.stats em uma nova célula:

```python
# Importando ttest_1samp
from scipy.stats import ttest_1samp
```

```python
# Realiza o teste t unilateral para uma amostra
stats, p_valor = ttest_1samp(tempo_resposta, 30, alternative = "less")
```

```python
print(p_valor)
```

###Conclusão
Vamos recapitular o que fizemos até o momento? Executamos um teste T, que é mais apropriado para essa situação, pois não conhecemos informações sobre a população, isto é, sobre todos os atendimentos de suportes realizados na empresa de software. Além disso, temos uma amostra pequena. Por isso, seguindo esses dois aspectos, o teste T é mais apropriado.
